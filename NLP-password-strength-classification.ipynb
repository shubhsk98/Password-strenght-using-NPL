# apply TF-IDS on our data

##### when we applu this on our data, u hv to make sure that data that u r going to pass thrugh ur TF-IDF that must be form characters.


```python
def word_divide_char(inputs):
    character=[]
    for i in inputs:
        character.append(i)
    return character
```

```python
word_divide_char('kzde5577')
```

```python
# so now very first pass this data into ur IDF 

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer(tokenizer=word_divide_char)
```

```python
X= vectorizer.fit_transform(x)
```

```python
X.shape
```

```python
vectorizer.get_feature_names()
```

```python
first_document_vector=X[0]
first_document_vector
```

```python
first_document_vector.T.todense()
```

```python
# now def dataframe for this
df=pd.DataFrame(first_document_vector.T.todense(),index=vectorizer.get_feature_names(),columns=['TF-IDF'])
```

```python
df
```

```python
df.sort_values(by=['TF-IDF'],ascending=False)
```

# ML-model

```python
logistic regression- classification applly on our model
```

```python
# firstly split our data

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
# now ur data ready for modeling ur model
```

```python
# import logostic regression class

from sklearn.linear_model import LogisticRegression
```

```python
# Apply Logistic on data as use-cas is Classification
```

```python
clf=LogisticRegression(random_state=0,multi_class='multinomial')
```

```python
# fit our data
clf.fit(X_train,y_train)
```

# doing prediction for specific custom data

```python
# im going to consider some random password. so i hv to just convert password in the form of array

dt=np.array(['%@123abcd'])
pred=vectorizer.transform(dt)
clf.predict(pred)
```

```python
above its means of avg strength password in similar way, u can do prediciton on test data
```

# doing prediction on X-Test data

```python
y_pred=clf.predict(X_test)
y_pred
```

# check Accuracy of your model using confusion_matrix,accuracy_score¶

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
```

# create report of your model

```python
# similar way we check classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
